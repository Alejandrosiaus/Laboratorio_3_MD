## estadistica desciptiva

```python
df.describe()
```

```python
#analisis de variables flotantes
missing_values = df.isnull().sum().sort_values(ascending=False)

missing_values.head(20)
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.isnull(), cbar=False)
plt.title("Mapa de valores faltantes")
plt.show()
```

## descripcion de variables respuestas


```python
plt.figure(figsize=(8,5))
sns.histplot(df["SalePrice"], kde=True)

plt.title("Distribución del precio de venta de las casas")
plt.xlabel("Precio de venta")
plt.ylabel("Frecuencia")

plt.show()
```

## analicis de correlacion

```python
correlation_matrix = df.corr(numeric_only=True)

plt.figure(figsize=(12,8))
sns.heatmap(correlation_matrix, cmap="coolwarm")

plt.title("Matriz de correlación entre variables")
plt.show()
```

```python
correlation_with_price = correlation_matrix["SalePrice"].sort_values(ascending=False)

correlation_with_price.head(10)
```

```python
# analicis de variables con el precio
plt.figure(figsize=(8,5))
sns.scatterplot(x="GrLivArea", y="SalePrice", data=df)

plt.title("Relación entre área habitable y precio")
plt.xlabel("Área habitable")
plt.ylabel("Precio de venta")

plt.show()
```
